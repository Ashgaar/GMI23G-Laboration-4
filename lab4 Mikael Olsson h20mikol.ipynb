Finally I setllted on using the bionomial theorem after some reading and this was by far the easiest way to do it. If you have (a+b)^n then lets say n = 4, you can then rewrite that and if you ignore the a's and the b's you get 1 4 6 4 1 which is what the 4th row on the pascal triangle should be.

NOTE!: The reason my explanation was so long is because I tried really hard to solve this, for me this was one of the hardest but most fun problems to solve.

```python
def pascalTriangle(n):
    for i in range(n):
        print([binomial(i, l) for l in range(i+1)])
        
pascalTriangle(10)
```
